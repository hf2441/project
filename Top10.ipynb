```python
import pandas as pd
import numpy as np
from datetime import time


df = pd.read_csv(r'/content/311_Service_Requests_2020.csv')

print (df)
```

```python
df.head()
df["Created Date"] = pd.to_datetime(df["Created Date"])
df["year"] = df["Created Date"].dt.year
```

```python
print(df["year"])
zip_10027 = df[(df["Incident Zip"] == 10027) & (df["year"] == 2020)]
```

```python
zip_10027.drop_duplicates(subset = "Complaint Type")
```

```python
top10 = zip_10027['Complaint Type']
hash = {}
elements = set(top10)

for i in elements:
  hash[i] = len(zip_10027[zip_10027["Complaint Type"] == i])
top10keys = sorted(hash, key = hash.get, reverse = True)[:10]

values = {i:hash[i] for i in top10keys}
values = pd.Series(values)
values
```
